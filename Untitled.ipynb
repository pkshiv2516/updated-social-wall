```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
import os

def screenshot_linkedin_post(post_url, output_path):
    
    try:
        output_path = os.path.join('sshts',output_path)
        # Navigate to the post
        print(f"Navigating to: {post_url}")
        driver.get(post_url)
        
        # Wait for page to load
        time.sleep(5)
        
        # Try to find the main post container
        post_selectors = [
            ".facebook",
            ".feed-shared-update-v2",
            ".update-components-actor--with-control-menu",
            "[data-test-app-aware-link]"
        ]
        
        post_element = None
        for selector in post_selectors:
            try:
                elements = driver.find_elements(By.CSS_SELECTOR, selector)
                if elements:
                    post_element = elements[0]
                    print(f"Found post using selector: {selector}")
                    break
            except:
                continue
        
        if post_element:
            # Scroll to the element
            # driver.execute_script("arguments[0].scrollIntoView();", post_element)
            # time.sleep(2)
            
            # Take screenshot of the element
            post_element.screenshot(output_path)
            print(f"Screenshot saved to: {output_path}")
            return True
        else:
            # Fallback: take full page screenshot
            print("Could not find specific post element, taking full page screenshot")
            driver.save_screenshot(output_path)
            return True
            
    except Exception as e:
        print(f"Error: {str(e)}")
        return False
        


# Usage
if __name__ == "__main__":
    import pandas as pd
    df = pd.read_csv('social_posts.csv')
    for i,post_url in enumerate(df['url']):
        success = screenshot_linkedin_post(post_url, f"my_linkedin_post{i}.png")
        
        if success:
            print("✅ Screenshot completed successfully!")
        else:
            print("❌ Screenshot failed!")
```

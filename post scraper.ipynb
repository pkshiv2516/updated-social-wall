```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import time

# Setup Chrome options for mobile emulation
mobile_emulation = {
    "deviceMetrics": { "width": 375, "height": 812, "pixelRatio": 3.0 },  # iPhone X
    "userAgent": (
        "Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) "
        "AppleWebKit/605.1.15 (KHTML, like Gecko) "
        "Version/13.0.3 Mobile/15E148 Safari/604.1"
    )
}

chrome_options = Options()
chrome_options.add_experimental_option("mobileEmulation", mobile_emulation)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-notifications")

# Initialize the driver
driver = webdriver.Chrome(options=chrome_options)

# Loop through post URLs
for i, post_url in enumerate(post_urls):
    output_path = os.path.join("sshts", "instagram", f"my_instagram_post{i}.png")
    print(f"Navigating to: {post_url}")
    driver.get(post_url)

    try:
        # Wait for the main article element to load
        element = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "article"))
        )
        # Wait a bit more for content like images to load
        time.sleep(3)
        # Capture screenshot of the article element
        element.screenshot(output_path)
        print(f"✅ Screenshot saved: {output_path}")
    except Exception as e:
        print(f"❌ Failed to capture post: {e}")

# Cleanup
driver.quit()
```

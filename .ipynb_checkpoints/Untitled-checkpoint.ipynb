```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load CSV
df = pd.read_csv("social_posts.csv")

# Function to extract tweet details from embed URL
def extract_tweet_data(embed_url):
    try:
        response = requests.get(embed_url)
        soup = BeautifulSoup(response.text, "html.parser")
        
        tweet_text = soup.find("p").text if soup.find("p") else "N/A"
        author = soup.find("span", {"class": "TweetAuthor-name"}).text if soup.find("span", {"class": "TweetAuthor-name"}) else "N/A"
        
        return {
            "text": tweet_text,
            "author": author
        }
    except Exception as e:
        return {
            "text": "Error",
            "author": str(e)
        }

# Apply extraction to first few rows (for demo)
results = df.head(5)["url"].apply(extract_tweet_data)
extracted_df = pd.DataFrame(results.tolist())
final_df = pd.concat([df.head(5).reset_index(drop=True), extracted_df], axis=1)

print(final_df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def fetch_og_data(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
        }
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')

        def get_meta(property_name):
            tag = soup.find('meta', property=property_name)
            return tag['content'] if tag and tag.has_attr('content') else ''

        return {
            'url': url,
            'title': get_meta('og:title'),
            'description': get_meta('og:description'),
            'image': get_meta('og:image'),
            'site_name': get_meta('og:site_name')
        }

    except Exception as e:
        print(f"Failed to fetch: {url} -> {e}")
        return {
            'url': url,
            'title': '',
            'description': '',
            'image': '',
            'site_name': '',
            'error': str(e)
        }

# Load CSV with post URLs
csv_path = 'social_posts.csv'
df = pd.read_csv(csv_path)

# Extract OG metadata for each URL
results = [fetch_og_data(url) for url in df['url']]

# Merge with original CSV
og_df = pd.DataFrame(results)
merged_df = pd.concat([df, og_df.drop(columns='url')], axis=1)

# Save results
output_csv = f'post_metadata_{datetime.now().strftime("%Y%m%d")}.csv'
merged_df.to_csv(output_csv, index=False)

print(f"âœ… Metadata extracted and saved to {output_csv}")
```
